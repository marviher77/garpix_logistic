### Извлекаем cargoes из data

```python
cargoes = pd.DataFrame(list(data[0]))
cargoes
```

```python
info = pd.DataFrame(list(cargoes['info']))
info
```

```python
margin = pd.DataFrame(list(cargoes['margin']))
margin.rename(columns = {'width':'margin_width', 'height':'margin_height', 'length':'margin_length'}, inplace = True )
full_cargoes = info.join(margin, on=None, how='left', sort=False)
full_cargoes
```

```python
size = pd.DataFrame(list(cargoes['size']))
size.rename(columns = {'width':'size_width', 'height':'size_height', 'length':'size_length'}, inplace = True )
full_cargoes = full_cargoes.join(size, on=None, how='left', sort=False)
full_cargoes
```

```python
full_cargoes = full_cargoes.join(cargoes[['mass', 'sort', 'type', 'count', 'cargo_id', 'stacking', 'turnover', 'stacking_limit']], on=None, how='left', sort=False)
full_cargoes.head()
```
